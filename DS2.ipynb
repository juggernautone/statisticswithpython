#### Coefficient of variation
-  the coefficient of variation (CV), also known as relative standard deviation (RSD), It is often expressed as a percentage. 
-  is defined as the ratio of the standard deviation  to the mean: $\frac{s}{\bar{x}}$ in a sample and $\frac{\sigma}{\mu}$ in the population.

**Example:**  I invest in stock "ISOM" and stock "FIN" for 5 years. the yearly close price of "ISOM"  is  (100,95,105, 105,95), and the yearly close price of "FIN" is (100,100,90,90,95). Then which stock is more risky?

```python
ISOM=pd.Series((100,95,105, 105,95))
FIN=pd.Series((100,100,90,90,95))
plt.plot(ISOM,label="ISOM")
plt.plot(FIN,label="FIN")
```

```python
ISOM_CV=ISOM.std(ddof=1)/ISOM.mean()
FIN_CV=FIN.std(ddof=1)/FIN.mean()
print "ISOM_CV=", ISOM_CV
print "FIN_CV=", FIN_CV
```

### Outliers
- An outlier is an observation that lies an abnormal distance from other values in a  sample 
- This definition does not help to decide what will be considered abnormal. Hence  it is necessary to characterize normal observations before we can pick abnormal observation.
- In most cases, outliers are **Good**


#### z-score and detecting ourliers
- z-score is used to justtify an observation to be outlier if the population  distribution is (almost) symmetric.
- z-score is define by $z-score=\frac{x_o-\bar{x}}{s}$
- if the distribution is not symmetric, z-score is not a good criterion.


**Example:** determine if $x_o=50$ is an outlier in money_change2?

```python
plt.figure(figsize=(30,10))
plt.hist(money_change2.dropna(),bins=20)
plt.plot(50,0,'or',markersize=30)
```
